##Predictions

###Choix du modele

la valeur cible est continue donc les modeles de regression

```python
from ast import mod
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import ElasticNet

models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42),
    'ElasticNet': ElasticNet()
}

predictions = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    predictions[name] = y_pred
```

###Validations

```python
from sklearn.metrics import mean_squared_error, r2_score
results = {}

for name, y_pred in predictions.items():
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'R²': r2}

# Affichage des résultats
for name, metrics in results.items():
    print(f"{name} - MSE: {metrics['MSE']:.4f}, R²: {metrics['R²']:.4f}")
```

la regression linanaire nous offre les meilleur resultat avec la plus bas MSE et le r_score la plus elever

```python
import pandas as pd
class ChatBot:
    def __init__(self, model, scaler):
        self.model = model
        self.scaler = scaler
        self.student_profile = {}

    def get_student_profile(self, name):
        questions = [
            ("abscences", "combien d'absences avez-vous eu dans les 3 derniers mois ?"),
            ("grade", "vous êtes en école, collège, lycée ou université ? (0,1,2,3,4 respectivement)", [0, 4]),
            ("participation", "à quel point participez-vous activement en classe sur une échelle de 1 à 5 ?", [1, 5]),
            ("gpa", "quelle est votre note moyenne actuelle ?", [0,20]),
            ("improvement", "avez-vous remarqué une amélioration ou une baisse de vos notes par rapport au semestre précédent ? (1 pour oui, 0 pour non)",  [0, 1]),
            ("study_week", "combien d'heures par semaine consacrez-vous à l'étude en dehors des cours ?"),
            ("revision", "combien de jours à l'avance commencez-vous à réviser pour un examen important ?"),
            ("someil", "combien d'heures de sommeil avez-vous dans votre journée ?", [0, 12]),
            ("stress", "comment évaluez-vous votre niveau de stress avant les examens sur une échelle de 1 à 5 ?", [1, 5]),
            ("motivation", "comment évaluez-vous votre niveau de motivation pour améliorer vos résultats sur une échelle de 1 à 5 ?", [1, 5]),
            ("ressource", "utilisez-vous souvent les ressources académiques fournies par le professeur ou cherchez-vous des supports supplémentaires ailleurs ? (1 pour oui, 0 pour non)",  [0, 1])
        ]

        for key, question, *range in questions:
            self.student_profile[key] = self.input_validation(input(f"{name}, {question}"), range)

        self.student_profile['gpa'] = float(self.student_profile['gpa'])

    def data_process(self):
        print(self.student_profile)
        data = {
            'Age': 20,
            'Gender': 1,
            'Ethnicity': 1,
            'ParentalEducation': 2,
            'StudyTimeWeekly': self.student_profile['study_week'],
            'Absences': self.student_profile['abscences'],
            'Tutoring': 1,
            'ParentalSupport' : 2,
            'Extracurricular': 1,
            'Sports': 0,
            'Music': 0,
            'Volunteering': 0,
            'GradeClass': self.student_profile['grade'],
            'GPA': 2.3
        }
        data_df = pd.DataFrame([data])
        numeric = ['Age', 'StudyTimeWeekly', 'Absences', 'GPA']
        data_df[numeric] = self.scaler.transform(data_df[numeric])
        return data_df

    def predict_gpa(self, data_proc):
        data_proc = data_proc.drop('GPA', axis=1)
        prediction = self.model.predict(data_proc)
        besoin = ['Age', 'StudyTimeWeekly', 'Absences']
        descale = data_proc[besoin]
        descale = descale.copy()
        descale.loc[:, 'GPA'] = prediction[0]
        descale = self.scaler.inverse_transform(descale)
        note_predit = descale[0][3] *5
        note_predit = round(note_predit, 2)
        return note_predit

    def generate_recommendations(self):
        recommendations = []

        conditions = {
            'abscences': (lambda: self.student_profile['abscences'] > 10, "Il faut essayer d'être moins absent."),
            'stress_revision': (lambda: self.student_profile['stress'] > 2 and self.student_profile['revision'] < 15, "Il faut mieux préparer à l'avance vos examens."),
            'someil': (lambda: self.student_profile['someil'] < 7, "Il faut plus de sommeil."),
            'stress_someil': (lambda: self.student_profile['someil'] < 7 and self.student_profile['stress'] > 2, "Le sommeil aide à gérer le stress."),
            'parreuse': (lambda: self.student_profile['someil'] > 10, "Dormir moins te donnera plus de temps pour tes autres activités."),
            'motivation': (lambda: self.student_profile['motivation'] < 4 and self.student_profile['revision'] < 10, "Il faut plus de motivation, fais-toi des petits systèmes de récompense pendant tes révisions."),
            'GradeClass_0': (lambda: self.student_profile['grade'] == 0, [
                (lambda: self.student_profile['someil'] < 7, "Essaye de mieux dormir, tu es en pleine croissance."),
                (lambda: self.student_profile['gpa'] < 3, "Prends un tuteur privé si possible."),
                "Fais du sport et/ou de la musique."
            ]),
            'GradeClass_4': (lambda: self.student_profile['grade'] == 4 and self.student_profile['gpa'] < 1 and self.student_profile['study_week'] < 10, "Pour les personnes à l'université, étudier +10h par semaine pourrait t'aider."),
            'improvement': (lambda: self.student_profile['improvement'] == 1, "De plus, continue ta méthode qui a conduit à ces ameliorations passer.")
        }

        for key, (verification, message) in conditions.items():
            if isinstance(verification, type(lambda: 0)):
                if verification():
                    if isinstance(message, list):
                        for condition, msg in message:
                            if condition():
                                recommendations.append(msg)
                    else:
                        recommendations.append(message)

        return recommendations

    def input_validation(self, user_input, valid_range=None):
        while True:
            try:
                user_input = int(user_input)
                if valid_range and (user_input < valid_range[0][0] or user_input > valid_range[0][1]):
                    print(f"Veuillez entrer un nombre entre {valid_range[0][0]} et {valid_range[0][1]}.")
                else:
                    return user_input
            except ValueError:
                print("Veuillez entrer un nombre entier.")
            user_input = input("Réessayez : ")

    def start_chat(self):
        name = input("Bonjour ! Je suis ici pour vous aider à améliorer vos résultats académiques. Comment vous appelez-vous ? ")
        self.get_student_profile(name)
        recommendations = self.generate_recommendations()
        data_prep = self.data_process()
        predicted_gpa = self.predict_gpa(data_prep)
        print("Voici nos prédictions sur votre future note :", predicted_gpa)
        print("Voici quelques suggestions pour améliorer vos résultats :")
        for recommendation in recommendations:
            print("- " + recommendation)

```
